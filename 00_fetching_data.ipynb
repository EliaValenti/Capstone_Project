## IMPORTAZIONE LIBRERIE UTILI

```python
import pandas as pd
import requests 
import os 
 
from selenium import webdriver
from selenium.webdriver.common.by import By 
```

## WEB SCRAPING SITUAS

```python
#guardo la risposta del server
url_situas = "https://situas.istat.it/web/#/territorio/body?id=74&dateFrom=2020-12-31"

resp_situas = requests.get (url_situas)

print(resp_situas.status_code)
```

Scarico il .csv 

```python
download_folder = os.path.abspath("fetching_data")

browser_options = webdriver.ChromeOptions()
download_pref = {"download.default_directory": download_folder, 
            "download.prompt_for_download": False} #così salva in automatico il file, no popup
browser_options.add_experimental_option("prefs", download_pref)

driver = webdriver.Chrome(options=browser_options) 
driver.get(url_situas) 

#dico di trovare l' elemento tramite ID nell' HTML e cliccarci su
driver.find_element(By.ID, "dati-report-export-btn").click() 
#dico di trovare il percorso di CSV tramite xpath e cliccare
#N.B. avendo messo "False" prima, il download partirà in automatico
driver.find_element(By.XPATH, "//span[text()='CSV']").click() 
```

```python
#creo la variabile che avrà come risultato il/i file/s che ho scaricato nella cartella fetching_data
raw_situas=os.listdir("fetching_data")
```
